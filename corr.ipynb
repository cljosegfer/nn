```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from tqdm import tqdm
from scipy.stats import pearsonr, kendalltau, spearmanr
from sklearn import metrics

from models import kNN, sNN, ggNN, sggNN
```

```python
H_train = torch.load('data/H_train.pt')
y_train = torch.load('data/y_train.pt')
s_train = torch.load('data/s_train.pt')

H_test = torch.load('data/H_test.pt')
y_test = torch.load('data/y_test.pt')
s_test = torch.load('data/s_test.pt')
```

# knn

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))

model = kNN(k = 94)
model.fit(H_train, y_train)
q_pearson = model.q(H_test, y_test)
axes[0].scatter(q_pearson, s_test);
axes[0].title.set_text(pearsonr(q_pearson, s_test)[0])

model = kNN(k = 98)
model.fit(H_train, y_train)
q_kendall = model.q(H_test, y_test)
axes[1].scatter(q_kendall, s_test);
axes[1].title.set_text(kendalltau(q_kendall, s_test)[0])

model = kNN(k = 113)
model.fit(H_train, y_train)
q_spearman = model.q(H_test, y_test)
axes[2].scatter(q_spearman, s_test);
axes[2].title.set_text(spearmanr(q_spearman, s_test)[0])

pearsonr(q_pearson, s_test)[0], kendalltau(q_kendall, s_test)[0], spearmanr(q_spearman, s_test)[0]
```
