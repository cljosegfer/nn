```python
import torch
import matplotlib.pyplot as plt
```

```python
# !pip install opentsne
from openTSNE import TSNE
```

```python
H_train = torch.load('data/H_train.pt')
y_train = torch.load('data/y_train.pt')
H_test = torch.load('data/H_test.pt')
y_test = torch.load('data/y_test.pt')
```

```python
H_train.shape, y_train.shape, H_test.shape, y_test.shape
```

```python
tsne = TSNE()
T_train = tsne.fit(H_train)
T_test = T_train.transform(H_test)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].scatter(T_train[:, 0], T_train[:, 1], c = y_train);
axes[0].title.set_text('train')
axes[1].scatter(T_test[:, 0], T_test[:, 1], c = y_test);
axes[1].title.set_text('test')
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
model.fit(H_train, y_train)
model.score(H_test, y_test)
```
