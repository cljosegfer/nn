```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from tqdm import tqdm

from models import kNN, sNN, ggNN
```

```python
H_train = torch.load('data/H_train.pt')
y_train = torch.load('data/y_train.pt')
yhat_train = torch.load('data/yhat_train.pt')
s_train = torch.load('data/s_train.pt')

H_test = torch.load('data/H_test.pt')
y_test = torch.load('data/y_test.pt')
yhat_test = torch.load('data/yhat_test.pt')
s_test = torch.load('data/s_test.pt')
```

```python
# grid = [1, 2, 3, 4, 5]
grid = np.linspace(start = 1, stop = 100, num = 60, dtype = int)
qs = []
for param in grid:
    model = kNN(k = param)
    model.fit(H_train, y_train)
    qs.append(model.q_mean(H_test, y_test))
```

```python
plt.scatter(grid, qs);
```

```python
# grid = [1, 2, 3, 4, 5]
grid = np.logspace(start = -2, stop = 2, num = 30)
qs = []
for param in grid:
    model = sNN(tau = param)
    model.fit(H_train, y_train)
    qs.append(model.q_mean(H_test, y_test))
```

```python
plt.scatter(grid, qs);
plt.xscale('log');
```
